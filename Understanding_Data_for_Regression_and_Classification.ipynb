# **Programming Best Practices / Coding Standards**

1. Use Meaningful Variable and Function Names. These
names should clearly describe the purpose (e.g., totalMarks instead of tm).

2. Follow Standard Coding Conventions i.e, language-specific standards such as PEP 8 for Python.

3. Write Clean and Readable Code.
It should be easy to understand, properly formatted, and well-indented.

4. Use Comments and Documentation.
Add comments where necessary and use docstrings to explain functions and classes.

5. Avoid Code Duplication.
Reuse code using functions, classes, or libraries instead of repeating logic.

6. Modularize the program.
Break large programs into smaller, manageable functions or modules.

7. Handle errors and exceptions properly, use exception handling to prevent program crashes and ensure smooth execution.

8. Avoid hard-coding values, use constants or configuration files instead of fixed values in the code.

9. Validate inputs, always check user input to avoid incorrect data and runtime errors.

10. Write efficient and optimized code.

 11. Use Version Control Systems, use tools like Git to track changes and collaborate effectively.

 12. Test the code thoroughly, perform unit testing and integration testing to ensure reliability.

13. Maintain consistent formatting, use consistent spacing, indentation, and braces throughout the code.

14. Remove Unused Code and Variables to improve readability and maintenance.

15. Write code that is easy to modify and extend in the future.

# **Understanding Datasets for Regression and Classification**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **REGRESSION DATASET**

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df_reg = pd.read_csv(url)

df_reg.head()
```

```python
df_reg.shape
```

```python
df_reg.info()
```

```python
df_reg.describe()
```

Target Variable: medv
- Represents median house value
- Continuous numeric data  
â†’ This is a Regression problem


```python
df_reg.isnull().sum()
```

```python
plt.hist(df_reg['medv'])
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.title("Distribution of Target Variable")
plt.show()
```

```python
df_reg.corr()['medv'].sort_values(ascending=False)
```

### **Regression Models**

- Target variable is continuous
- Linear relationships exist between features and target
- No class labels involved

**Suitable Models:**
- Linear Regression
- Decision Tree
- Random Forest


# **CLASSIFICATION DATASET**


```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df_clf = pd.read_csv(url)

df_clf.head()
```

```python
df_clf.shape
```

```python
df_clf.info()
```

```python
df_clf.describe()
```

Target Variable: species
- Categorical (Setosa, Versicolor, Virginica)
â†’ This is a Classification problem

```python
df_clf['species'].value_counts()
```

```python
pd.plotting.scatter_matrix(
    df_clf.iloc[:, :-1],
    figsize=(8, 8)
)
plt.show()
```

### **Classification Models**

- Target variable is categorical
- Classes are clearly separable
- Balanced class distribution

**Suitable Models:**
- Logistic Regression
- K-Nearest Neighbors (KNN)
- Decision Tree Classifier
- Support Vector Machine (SVM)


## **Conclusion**

- Regression dataset has a continuous target, so regression models are suitable.
- Classification dataset has discrete class labels, so classification models are suitable.
- Understanding data distribution, relationships, and target type is essential before model selection.
