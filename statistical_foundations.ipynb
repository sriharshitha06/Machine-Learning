```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate a base dataset
np.random.seed(42) # for reproducibility
data_without_outliers = np.random.normal(loc=50, scale=10, size=1000)

# 2. Introduce outliers to this dataset
outliers = np.array([1, 2, 3, 100, 110, 120]) # A few extremely low and high values
data_with_outliers = np.concatenate((data_without_outliers, outliers))

# 3. Calculate mean and median for data_without_outliers
mean_without_outliers = np.mean(data_without_outliers)
median_without_outliers = np.median(data_without_outliers)

print(f"--- Dataset without Outliers ---")
print(f"  Mean: {mean_without_outliers:.2f}")
print(f"  Median: {median_without_outliers:.2f}")

# 4. Calculate mean and median for data_with_outliers
mean_with_outliers = np.mean(data_with_outliers)
median_with_outliers = np.median(data_with_outliers)

print(f"\n--- Dataset with Outliers ---")
print(f"  Mean: {mean_with_outliers:.2f}")
print(f"  Median: {median_with_outliers:.2f}")

# 5. Compare the mean and median values to observe the effect of outliers
print(f"\n--- Comparison ---")
print(f"Observation: The mean changed from {mean_without_outliers:.2f} to {mean_with_outliers:.2f}, while the median changed from {median_without_outliers:.2f} to {median_with_outliers:.2f}.")
print(f"The mean is more sensitive to outliers, being pulled towards the extreme values, whereas the median remains relatively stable.")

# 6, 7, 8, 9. Create a visualization to display the distribution of data_with_outliers
plt.figure(figsize=(10, 6))
sns.histplot(data_with_outliers, bins=50, kde=True, color='skyblue', edgecolor='black')

# Mark the mean and median
plt.axvline(mean_with_outliers, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_with_outliers:.2f}')
plt.axvline(median_with_outliers, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_with_outliers:.2f}')

plt.title('Distribution with Outliers: Mean vs. Median', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate a population dataset
population_data = np.random.normal(loc=100, scale=15, size=10000)

# 2. Set sample_size and num_samples
sample_size = 30
num_samples = 1000

# 3. Create an empty list named sample_means
sample_means = []

# 4. Loop num_samples times to take random samples and calculate their means
for _ in range(num_samples):
    # a. Take a random sample of sample_size from population_data
    sample = np.random.choice(population_data, size=sample_size, replace=False)
    # b. Calculate the mean of the current sample
    current_sample_mean = np.mean(sample)
    # c. Append the calculated sample mean to the sample_means list
    sample_means.append(current_sample_mean)

# 5. Calculate the population mean
population_mean = np.mean(population_data)

# 6. Calculate the mean of all the values in the sample_means list
mean_of_sample_means = np.mean(sample_means)

# 7. Print population_mean and mean_of_sample_means
print(f"Population Mean: {population_mean:.4f}")
print(f"Mean of Sample Means: {mean_of_sample_means:.4f}")

# 8. Create a histogram of the sample_means with KDE
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, bins=30, kde=True, color='skyblue', edgecolor='black')

# 9. Add vertical lines for population_mean and mean_of_sample_means
plt.axvline(population_mean, color='red', linestyle='dashed', linewidth=2, label=f'Population Mean: {population_mean:.2f}')
plt.axvline(mean_of_sample_means, color='green', linestyle='dashed', linewidth=2, label=f'Mean of Sample Means: {mean_of_sample_means:.2f}')

# 10. Add title, x-label, y-label, and a legend
plt.title('Distribution of Sample Means', fontsize=16)
plt.xlabel('Sample Mean', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# 11. Display the plot
plt.show()
```
