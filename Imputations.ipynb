```python
import pandas as pd
import numpy as np

print("Pandas library imported successfully.")

# Create a sample nominal DataFrame directly
data = {
    'Color': ['Red', 'Blue', 'Green', 'Red', None, 'Blue', 'Green'],
    'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small', 'Medium', None],
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C'],
    'Material': ['Cotton', 'Wool', 'Silk', 'Cotton', 'Wool', None, 'Silk']
}
nominal_df = pd.DataFrame(data)

# Display the first 5 rows of the DataFrame
print("First 5 rows of the nominal data (with missing values for imputation):")
print(nominal_df.head())
```

```python
data_numerical = {
    'Age': [25, 30, None, 40, 35, 28, 45],
    'Height': [170.5, 175.0, 180.2, 165.1, None, 172.8, 178.0],
    'Weight': [65, 70, 75, None, 68, 62, 80],
    'Score': [85, 92, 78, 95, 88, None, 90]
}
numerical_df = pd.DataFrame(data_numerical)

print("First 5 rows of the numerical data (with missing values for imputation):")
print(numerical_df.head())
```

```python
import pandas as pd

# Recreate the numerical DataFrame from original data for mean imputation demonstration
data_numerical_for_mean_imputation = {
    'Age': [25, 30, None, 40, 35, 28, 45],
    'Height': [170.5, 175.0, 180.2, 165.1, None, 172.8, 178.0],
    'Weight': [65, 70, 75, None, 68, 62, 80],
    'Score': [85, 92, 78, 95, 88, None, 90]
}
mean_imputed_numerical_df = pd.DataFrame(data_numerical_for_mean_imputation)

print("Original numerical data (before mean imputation):\n")
print(mean_imputed_numerical_df.head())
print("\nMissing values before mean imputation:\n")
print(mean_imputed_numerical_df.isnull().sum())
```

```python
for col in mean_imputed_numerical_df.columns:
    if mean_imputed_numerical_df[col].isnull().any():
        mean_value = mean_imputed_numerical_df[col].mean()
        mean_imputed_numerical_df[col] = mean_imputed_numerical_df[col].fillna(mean_value)

print("\nFirst 5 rows of numerical_df after mean imputation:\n")
print(mean_imputed_numerical_df.head())

print("\nMissing values in numerical_df after mean imputation:\n")
print(mean_imputed_numerical_df.isnull().sum())
```

```python
for col in nominal_df.columns:
    if nominal_df[col].isnull().any():
        mode_value = nominal_df[col].mode()[0]
        nominal_df[col].fillna(mode_value, inplace=True)

print("First 5 rows of nominal_df after mode imputation:")
print(nominal_df.head())

print("\nMissing values in nominal_df after imputation:")
print(nominal_df.isnull().sum())
```

```python
for col in nominal_df.columns:
    if nominal_df[col].isnull().any():
        mode_value = nominal_df[col].mode()[0]
        nominal_df[col] = nominal_df[col].fillna(mode_value)

print("First 5 rows of nominal_df after mode imputation:")
print(nominal_df.head())

print("\nMissing values in nominal_df after imputation:")
print(nominal_df.isnull().sum())
```

```python
for col in numerical_df.columns:
    if numerical_df[col].isnull().any():
        median_value = numerical_df[col].median()
        numerical_df[col] = numerical_df[col].fillna(median_value)

print("First 5 rows of numerical_df after median imputation:")
print(numerical_df.head())

print("\nMissing values in numerical_df after imputation:")
print(numerical_df.isnull().sum())
```
