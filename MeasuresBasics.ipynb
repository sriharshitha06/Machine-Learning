```python
data =[65,78,82,91,74,88,67,69,90,85,73,76,84,77,80,72,79,95,88,66,70,83,78,89,93,81,87,77,69,74]
```

```python
n =  len(data)
n
```

```python
s = 0
for i in range(n):
  s = s + data[i]
s
```

```python
mean = s/n
print("Mean",mean)
```

```python
data.sort()
data
```

```python
if n % 2 == 1 :
 median = data[n//2]
else :
  mid1 = data[n//2-1]
  mid2 = data[n//2]
  median = (mid1 + mid2)/2
print("Median",median)
```

```python
counts = {}
for i in data:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1

max_freq = 0
for freq in counts.values():
    if freq > max_freq:
        max_freq = freq

mode = [key for key, value in counts.items() if value == max_freq]

if len(mode) == len(data):
    print("No mode (all values appear once)")
else:
    print("Mode:", mode)
```

```python
sum_squared_diff = 0
for x in data:
    sum_squared_diff += (x - mean) ** 2
variance = sum_squared_diff / n
print("Variance:", variance)
```

```python
sd = variance ** 0.5
print("Standard Deviation:", sd)
```
