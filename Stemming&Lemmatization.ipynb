```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
sample_sentence = "Right takeaways were preserved from oscar nominations."
tokenized_words = word_tokenize(sample_sentence)
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()
wordnet_lemmatizer = WordNetLemmatizer()
print(sample_sentence)
print(f"After tokenization: {tokenized_words}")
porter_stemmed_words = [porter_stemmer.stem(word) for word in tokenized_words]
print(f"Porter Stemmed: {porter_stemmed_words}")
lancaster_stemmed_words = [lancaster_stemmer.stem(word) for word in tokenized_words]
print(f"Lancaster Stemmed: {lancaster_stemmed_words}")
lemmatized = [wordnet_lemmatizer.lemmatize(word) for word in tokenized_words]
print(f"Lemmatized : {lemmatized}")
```
