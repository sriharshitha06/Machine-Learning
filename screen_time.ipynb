```python
import pandas as pd

file_path = '/content/screen-time-v2.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify categorical columns (object dtype, excluding 'User_ID')
categorical_cols = df.select_dtypes(include='object').columns.tolist()
if 'User_ID' in categorical_cols:
    categorical_cols.remove('User_ID')

print(f"Categorical columns identified for analysis: {categorical_cols}\n")

# Loop through each categorical column to get value counts and plot
for col in categorical_cols:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts())
    print("\n" + "-"*50 + "\n") # Separator for better readability

    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify categorical columns (object dtype, excluding 'User_ID')
categorical_cols = df.select_dtypes(include='object').columns.tolist()
if 'User_ID' in categorical_cols:
    categorical_cols.remove('User_ID')

print(f"Categorical columns identified for analysis: {categorical_cols}\n")

# Loop through each categorical column to get value counts and plot
for col in categorical_cols:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts())
    print("\n" + "-"*50 + "\n") # Separator for better readability

    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, hue=col, palette='viridis', legend=False)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns (int64 and float64 dtypes)
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"Numerical columns identified for analysis: {numerical_cols}\n")

# Loop through each numerical column to create a histogram
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=col, kde=True, bins=20) # kde=True for density curve, bins for better visualization
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```
