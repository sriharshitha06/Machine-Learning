```python
import pandas as pd
```

reading the file

```python
df=pd.read_csv('/content/data.csv')
df
```

Adding a column which will have Binary values 0 or 1, in accordance with 'diagonsis' column ,if M, then 1, if B then 0

```python
#diag_binary
x=[]
for i in df['diagnosis']:
  if i=='M':
    x.append(1)
  else:
    x.append(0)

df['diag_binary']=x
df
```

```python
df.columns
```

Making a Copy and removing categorical and non numerical columns from the copy

```python
df_new=df.copy()
df_new.drop(columns=['id','diagnosis','Unnamed: 32','diag_binary'],inplace=True)
df_new
```

```python
Y=df['diag_binary']
X=df_new
X
```

Import Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

Diving the dataset in Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
X_train
```

```python
Y_train
```

training the model

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

Imports for checking accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

Checking Accuracy

```python
acc=accuracy_score(Y_test,Y_pred)
acc
```

```python
cm=confusion_matrix(Y_test,Y_pred)
cm
```
