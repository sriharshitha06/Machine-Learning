## Linear Regression

```python
import numpy as np

# Sample dataset (x values and actual y values)
x = np.array([1, 2, 3, 4, 5])
y_actual = np.array([2, 4, 5, 4, 5])

# Three different model lines (y = mx + c)
def model1(x):
    return 0.8*x + 1

def model2(x):
    return 1.0*x + 0.5

def model3(x):
    return 0.5*x + 2

# Cost function (Mean Squared Error)
def cost_function(y_actual, y_pred):
    n = len(y_actual)
    cost = (1/(2*n)) * np.sum((y_pred - y_actual)**2)
    return cost

# Predictions
y_pred1 = model1(x)
y_pred2 = model2(x)
y_pred3 = model3(x)

# Cost calculations
cost1 = cost_function(y_actual, y_pred1)
cost2 = cost_function(y_actual, y_pred2)
cost3 = cost_function(y_actual, y_pred3)

print("Cost for Model 1:", cost1)
print("Cost for Model 2:", cost2)
print("Cost for Model 3:", cost3)

# Determine minimum cost model
costs = [cost1, cost2, cost3]
min_cost = min(costs)

best_model = costs.index(min_cost) + 1

print("\nMinimum Cost:", min_cost)
print("Best Model Line: Model", best_model)
```
