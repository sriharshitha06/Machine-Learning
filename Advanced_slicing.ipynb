List unpacking

```python
nums = [1, 2, 3, 4, 5]

head, *tail = nums
print("head:",head,"tail:",tail)
```

```python
head,*mid,end = [1,2,3,4]
print(head,mid,end)
```

```python
#function packing
def add(*numbers):
    print(numbers)

add(1, 2, 3, 4)
```

Map and Zip for lists

map() takes function+list

```python
def cube(n) : return n**3
nums = [5,6,7,8]
cubes = list(map(cube,nums))
print("Cubes:",cubes)
```

zip() pairs elements from multiple lists

```python
names = ["A", "B", "C"]
marks = [90, 85, 88]

combined = list(zip(names, marks))

print("zipped:",combined)
```
