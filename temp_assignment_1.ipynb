# Exercise: map() for Feature Transformation

You want to normalize features using a simple rule.

features = [10, 20, 30]

Task:
- Use map() with a lambda to divide each value by 10
- Store result in normalized_features
- Print normalized_features


```python
# Write your solution here
features = [10, 20, 30]
normalized_features = list(map(lambda x: x / 10, features))
print("normalized_features :",normalized_features)
```

# Exercise: Counter for Token Frequency

You are analyzing token frequency in generated text.

tokens = ["ai", "ml", "ai", "genai", "ml", "ai"]

Task:
- Use Counter to count token frequency
- Store result in token_counts
- Print token_counts


```python
# Write your solution here
from collections import Counter
tokens = ["ai", "ml", "ai", "genai", "ml", "ai"]
token_counts = Counter(tokens)
print("token_counts :",token_counts)
```
